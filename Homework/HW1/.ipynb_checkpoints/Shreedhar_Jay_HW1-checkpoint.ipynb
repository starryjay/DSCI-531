```python
happiness = pd.read_csv('data_la_happiness.csv')
happiness
```

# Question 1
## 1.1: Data Sanctity

```python
happiness.isna().sum()[happiness.isna().sum() > 0]
```

### Analysis
There are 5 missing values in `meanHSize` and 10 in `percent_bachelorPlus`. Since there are so few missing values, we can simply drop them.

```python
happiness = happiness.dropna()
happiness
```

```python
happiness.mean()[['meanHSize', 'percent_bachelorPlus', 'totalRace1', 'totalRace2', 'meanvalence']]
```

```python
happiness.median()[['meanHSize', 'percent_bachelorPlus', 'totalRace1', 'totalRace2', 'meanvalence']]
```

## 1.2: Outlier Detection

```python
sns.set_theme(rc={'figure.figsize':(4,8)})
sns.boxplot(happiness['meanvalence'], width=0.25, flierprops={"marker": "x"})
```

```python
q1 = happiness['meanvalence'].quantile(0.25)
q3 = happiness['meanvalence'].quantile(0.75)
iqr = q3 - q1
happiness = happiness.loc[(happiness['meanvalence'] >= (q1 - (1.5 * iqr))) & (happiness['meanvalence'] <= (q3 + (1.5 * iqr)))]
happiness
```

## 1.3: Variable Relationships

```python
sns.pairplot(happiness)
```

### Analysis

Let's ignore TRACT since it's just an index variable.  

`meanHSize` is moderately negatively correlated with `percent_bachelorPlus` and `totalRace1`, weakly negatively correlated with `meanvalence`, and moderately positively correlated with `totalRace2`. This means that the more people there are of race 2, the larger the household size tends to be. The inverse is true for the proportion of people with a Bachelor's degree or higher, and the number of people from race 1. As household size increases, happiness decreases a small amount.

`percent_bachelorPlus` is moderately negatively correlated with `totalRace2`, weakly positively correlated with `meanvalence`, and moderately positively correlated with `totalRace1`.  This means that the number of people from race 2 is negatively associated with the percentage of people with a Bachelor's degree or higher, the number of people from race 1 is positively associated with the same, and more people with Bachelor's degrees cause a small increase in happiness.  

`totalRace1` is moderately negatively correlated with `totalRace2` and weakly positively correlated with `meanValence`. Its distribution appears to be massively right-skewed, with the vast majority of values falling between 0 and roughly 200. This means there are not very many people of race 1, but their presence appears to discourage the presence of people from race 2 and is positively associated with increases in happiness.

The distribution of `totalRace2` is somewhat right-skewed, with most values being below roughly 2500. There are more neighborhoods with higher numbers of people from race 2 than from race 1 in the sample.

## 1.4: Simple Models and Residual Analysis

```python
y = happiness['meanvalence']
X1 = sm.add_constant(happiness['totalRace1'])
X2 = sm.add_constant(happiness['totalRace2'])
lm_race1 = sm.OLS(y, X1)
lm_race2 = sm.OLS(y, X2)
```

```python
results1 = lm_race1.fit()
results1.summary()
```

```python
fig, ax = plt.subplots(figsize=(10, 8))
ax.axhline(y=0, color='k', linewidth=1)
ax.set_title('Residual Plot for totalRace1:meanvalence OLS')
ax.set_xlabel('Fitted Values')
ax.set_ylabel('Residuals')
sns.scatterplot(x=results1.fittedvalues, y=results1.resid, s=5, color='midnightblue')
```

```python
results2 = lm_race2.fit()
results2.summary()
```

```python
fig, ax = plt.subplots(figsize=(10, 8))
ax.axhline(y=0, color='k', linewidth=1)
ax.set_title('Residual Plot for totalRace2:meanvalence OLS')
ax.set_xlabel('Fitted Values')
ax.set_ylabel('Residuals')
sns.scatterplot(x=results2.fittedvalues, y=results2.resid, s=5, color='midnightblue')
```

### Analysis


The residuals for each plot mirror each other - in the case of `totalRace1`, they range from roughly 5.78 to roughly 6.00, with the vast majority of values being clustered around 5.78. Residuals for `totalRace2` range from around 5.675 to 5.875, with most values being between 5.825 and 5.875. This violates the linear relationship assumption. The residuals plots appear to be mildly heteroskedastic, with the variance of the residuals being highest where the most points are and decreasing as they move further out.

## 1.5: Log Transformation

### Analysis
Due to the heteroskedasticity of the residuals plots and the violation of the linear relationship assumption, we must log-transform the data to ensure that nonlinearity in the data is removed.

```python
loglm_race1 = sm.OLS(y, sm.add_constant(np.log(happiness['totalRace1'])))
loglm_race2 = sm.OLS(y, sm.add_constant(np.log(happiness['totalRace2'])))
```

```python
logres1 = loglm_race1.fit()
logres2 = loglm_race2.fit()
```

```python
logres1.summary()
```

```python
fig, ax = plt.subplots(figsize=(10, 8))
ax.axhline(y=0, color='k', linewidth=1)
#ax.set_ylim(-12.5, 12.5)
sns.scatterplot(logres1.resid, s=5, color='midnightblue')
```

```python
fig, ax = plt.subplots(figsize=(10, 8))
ax.axhline(y=0, color='k', linewidth=1)
#ax.set_ylim(-12.5, 12.5)
sns.scatterplot(logres2.resid, s=5, color='midnightblue')
```

### Analysis
As evidenced by the above residuals plots, the log-transform did indeed address the nonlinear relationships in the data, making the residual plots homoskedastic as they should be.

# Question 2

## 2.1: Model Building

```python
X_multi = happiness.loc[:, ['percent_bachelorPlus', 'totalRace1', 'totalRace2']]
X_multi['totalRace1'] = np.log(X_multi['totalRace1'])
X_multi['totalRace2'] = np.log(X_multi['totalRace2'])
X_multi = sm.add_constant(X_multi)
multi_model = sm.OLS(y, X_multi)
multi_res = multi_model.fit()
multi_res.summary()
```

It appears that `percent_bachelorPlus` and `totalRace1` are positively related to `meanvalence`, with each 1-percent increase in `percent_bachelorPlus` causing a 0.0018-unit increase in `meanvalence`, and each 1-person increase in `totalRace1` causing a 0.0051-unit increase in `meanvalence`, holding the other variable constant. Essentially, as the percentage of the census tract population with a Bachelor's degree or higher and the number of people from race 1 increase, happiness also increases. However, as the number of people from race 2 increases, happiness _decreases_. For each 1-person decrease in the number of people from race 2, happiness increases by 0.0077 units.

## Question 2.2: Scatterplot Analysis

```python
fig, ax = plt.subplots(figsize=(10, 8))
sns.scatterplot(x=X_multi.iloc[:, 1], y=y, hue=X_multi.iloc[:, 3], palette="rocket", edgecolors='xkcd:mahogany', s=15)
```

```python
multi_res.predict()
```

```python
fig, ax = plt.subplots(figsize=(10, 8))
ax.set_ylabel('predicted meanvalence')
sns.scatterplot(x=X_multi.iloc[:, 1], y=multi_res.predict(), hue=X_multi.iloc[:, 3], palette="rocket", edgecolors='xkcd:mahogany', s=15)
```

### Analysis
It appears that the model predicted a strongly correlated increase in happiness as the percentage of people with a Bachelor's degree or higher went up, and also predicted a decrease in people from race 2. The reality was more complex, with some people from race 2 reporting high levels of happiness and having Bachelor's degrees or better. The association between a Bachelor's degree (or better) and happiness was also not as strong as predicted.

## Question 2.3: Correlation Heatmap

```python
mv = happiness['meanvalence'].reset_index(drop=True)
l1 = np.log(happiness['totalRace1']).reset_index(drop=True)
l2 = np.log(happiness['totalRace2']).reset_index(drop=True)
```

```python
data = pd.DataFrame([mv, multi_res.predict(), l1, l2]).T
data.columns = ['meanvalence', 'predicted_meanvalence', 'log(totalRace1)', 'log(totalRace2)']
```

```python
fig, ax = plt.subplots(figsize=(10, 8))
sns.heatmap(data.corr(), annot=True)
```

### Analysis 

It appears that `meanvalence` is moderately negatively correlated with `log(totalRace2)` and moderately positively correlated with `log(totalRace1)`. These correlations are strengthened twofold when looking at `predicted_meanvalence` instead. This means our model is in fact biased, as it predicted much stronger correlations between race and happiness than actually existed.

# Question 3

## Question 3.1: Protected Variable Analysis

```python
X_multi2 = happiness.loc[:, ['percent_bachelorPlus', 'totalRace1']].reset_index(drop=True)
X_multi2['totalRace1'] = np.log(X_multi2['totalRace1'])
multi_model2 = sm.OLS(mv, X_multi2)
multi_res2 = multi_model2.fit()
multi_res2.summary()
```

```python
data = pd.DataFrame([mv, multi_res2.predict(), l1, l2]).T
data.columns = ['meanvalence', 'predicted_meanvalence', 'log(totalRace1)', 'log(totalRace2)']
fig, ax = plt.subplots(figsize=(10, 8))
sns.heatmap(data.corr(), annot=True)
```

### Analysis

While the positive correlation between happiness and the number of people from race 1 was inflated almost threefold by this new model's predictions, the negative correlation between people of race 2 and happiness was actually deflated almost threefold. It appears that the bias was not reduced but simply shifted.
